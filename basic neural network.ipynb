```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #for 3d graphs
```

Generating random input data to train on:

```python
observations = 1000

xs = np.random.uniform(-10, 10, (observations, 1))
zs = np.random.uniform(-10, 10, (observations, 1))

inputs = np.column_stack((xs, zs))

inputs.shape
```

Creating targets to which we will aim at:

```python
noise = np.random.uniform(-10, 10, (observations, 1))

targets = 2*xs - 3*zs + 5 + noise

targets.shape
```

Plotting Training Data:

```python
targets = targets.reshape(observations, )
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
ax.plot(xs, zs, targets)
ax.set_xlabel("xs")
ax.set_ylabel("zs")
ax.set_zlabel("Targets")
ax.view_init(azim = 100)
plt.show()
targets = targets.reshape(observations, 1)
```

Initializing Varibles:

```python
init_range = 0.1

weights = np.random.uniform(-init_range, init_range, size = (2, 1))

biases = np.random.uniform(-init_range, init_range, size = 1)

print(weights)
print(biases)
```

Setting Learning Rate:

```python
learning_rate = 0.02
```

Training Model:

```python
for i in range(100):
    #calculating outputs:
    outputs = np.dot(inputs, weights) + biases
    deltas = outputs - targets
    
    #compare outputs to targets through the loss:
    loss = np.sum(deltas ** 2) / 2 / observations
    
    print(loss)
    
    deltas_scaled = deltas / observations
    
    #adjusting weights and biases:
    weights = weights - learning_rate * np.dot(inputs.T, deltas_scaled)
    biases = biases - learning_rate * np.sum(deltas_scaled)
```

Print weights and biases and see if we have worked correctly:

```python
print(weights, biases)
```

Plot last outputs vs targets:

```python
plt.plot(outputs, targets)
plt.xlabel("outputs")
plt.ylabel("targets")
plt.show()
```
